```python
simulation_directory = '/home/baparker/m12r_res7100/'
part = gizmo.io.Read.read_snapshots('all', 'redshift', 0.702, simulation_directory)
# Set simulation directory and read in all particle data
```

```python
gas_loc = part['gas']['position']
gas_temp = part['gas']['temperature']
gas_den = part['gas']['density']
```

```python
x = gas_loc[:, 0]
y = gas_loc[:, 2]
z = gas_loc[:, 1]

xoffset = 89990.81
yoffset = 89786.91
zoffset = 92491.32

length = 100

ind = (x < xoffset + length) & (x > xoffset - length) & (y < yoffset + length) & (y > yoffset - length)
#ind = (x==x)

x = x[ind] - xoffset
y = y[ind] - yoffset
z = z[ind] - zoffset

nH = part['gas'].prop('number.density')[ind] / (u.cm**3)

density = part['gas']['density'][ind]

#mass_frac = part['gas'].prop('massfraction.hydrogen')[::scale]
mass_frac = part['gas']['massfraction'][:,0][ind]

temp_plot = gas_temp[ind]

thickness = part['gas']['size'][ind]

metallicity_fe = part['gas'].prop('metallicity.fe')[ind]

gas_loc_plot = np.vstack((x,y)).T
```

```python
sort = np.argsort(mass_frac)
nH_plot, temp_plot, mass_frac = nH[sort], temp_plot[sort], mass_frac[sort]
plt.scatter(nH_plot[::100], temp_plot[::100], c=mass_frac[::100], s=1, cmap='spring', norm=LogNorm())
plt.colorbar(label = 'Mass Fraction')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number Density [cm$^{{-3}}$]')
plt.ylabel('Temperature [K]')
plt.title('Phase Diagram')
plt.savefig('Phase_Diagram.png', dpi=500)
```

```python
n = 200
gridx = np.linspace(x.min(), x.max(), n+1)
gridy = np.linspace(y.min(), y.max(), n+1)
```

```python
nH_sum, _, _ = binned_statistic_dd(gas_loc_plot, nH, bins=[gridx, gridy], statistic='sum')
nH_sum = np.nan_to_num(nH_sum, nan=nH_sum.min())
nH_sum[nH_sum == 0] = 1e-10
```

```python
fig, ax = plt.subplots(figsize=(6,4))
plt.imshow(nH_sum, norm = LogNorm(vmin=1e-4, vmax=1e4), cmap = 'magma', extent = [x.min(), x.max(), y.min(), y.max()], origin='lower')#vmin = 1e-8, vmax = 1e-1
plt.xlabel('Distance from Center (kpc h$^{{-1}}$)')
plt.ylabel('Distance from Center (kpc h$^{{-1}}$)')
plt.colorbar(label = r'Hydrogen Number Density (1/cm$^3$)')
plt.title('Binned Hydrogen Density')
plt.savefig('Hydrogen_Column_Density_Binned.png', dpi=500)
```

```python
temp_avg, _, _ = binned_statistic_dd(gas_loc_plot, temp_plot, bins=[gridx, gridy], statistic='mean')
temp_avg = np.nan_to_num(temp_avg, nan=1e0)
temp_avg[temp_avg == 0] = 1e0
```

```python
fig, ax = plt.subplots(figsize=(6,4))
plt.imshow(temp_avg, norm = LogNorm(vmin=1e4, vmax=1e6), cmap = 'hot', extent = [x.min(), x.max(), y.min(), y.max()], origin='lower')#vmin = 1e-8, vmax = 1e-1
plt.xlabel('Distance from Center (kpc h$^{{-1}}$)')
plt.ylabel('Distance from Center (kpc h$^{{-1}}$)')
plt.colorbar(label = r'Temperature (K)')
plt.title('Binned Gas Temperature')
plt.savefig('Temperature_Binned.png', dpi=500)
```

```python
kcpa = cosmo.kpc_comoving_per_arcmin(redshift)
```

```python
x_arcmin = x*u.kpc / kcpa
y_arcmin = y*u.kpc / kcpa
```

```python
x_arcsec = x_arcmin.to(u.arcsec)
y_arcsec = y_arcmin.to(u.arcsec)
```

```python
arcsec_pp = (x_arcsec.max() - x_arcsec.min()) / n
print(arcsec_pp)
```
