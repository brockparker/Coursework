### Gizmo Analysis

```python
gizmo.io
# Initialize Gizmo
```

```python
simulation_directory = '/home/baparker/m12r_res7100/'
part = gizmo.io.Read.read_snapshots('all', 'redshift', 0.702, simulation_directory)
# Set simulation directory and read in all particle data
```

```python
gas_loc = part['gas']['position']
gas_temp = part['gas']['temperature']
gas_den = part['gas']['density']
```

```python
x = gas_loc[:, 0]
y = gas_loc[:, 2]
z = gas_loc[:, 1]

xoffset = 89990.81
yoffset = 89786.91
zoffset = 92491.32

length = 100

ind = (x < xoffset + length) & (x > xoffset - length) & (y < yoffset + length) & (y > yoffset - length)
#ind = (x==x)

x = x[ind] - xoffset
y = y[ind] - yoffset
z = z[ind] - zoffset

nH = part['gas'].prop('number.density')[ind]

density = gas_den[ind]

#mass_frac = part['gas'].prop('massfraction.hydrogen')[::scale]
mass_frac = part['gas']['massfraction'][:,0][ind]

temp = gas_temp[ind]

metallicity_fe = part['gas'].prop('metallicity.fe')[ind]

gas_loc_plot = np.vstack((x,y)).T
```

## Plots

### Generate Phase Diagram

```python
sort = np.argsort(mass_frac)
nH_plot, temp_plot, mass_frac_plot = nH[sort], temp[sort], mass_frac[sort]
```

```python
plt.scatter(nH_plot, temp_plot, c=mass_frac_plot, s=1, cmap='spring', norm=LogNorm())
plt.colorbar(label = 'Mass Fraction')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number Density [cm$^{{-3}}$]')
plt.ylabel('Temperature [K]')
#plt.title('Phase Diagram')
plt.savefig('Phase_Diagram.png', dpi=500)
```

### Plot Interpolated Emissivity Table

```python
grid_file = '/home/baparker/code/c23.00/project/Lya.grid'
output_file = '/home/baparker/code/c23.00/project/Lya.emissivities'
```

```python
grid = pd.read_csv(grid_file, delimiter = '\t')
grid_dense = np.array(grid['HDEN=%f L'])
grid_temp = np.array(grid['CONSTANT '])
```

```python
output = pd.read_csv(output_file, delimiter = '\t')
grid_lya = np.log10(np.array(output['H  1 1215.67A']))
```

```python
den_min, den_max = grid_dense.min(), grid_dense.max()
temp_min, temp_max = grid_temp.min(), grid_temp.max()
xg, yg = np.mgrid[den_min:den_max:1000j, temp_min:temp_max:1000j]

interp = griddata((grid_dense, grid_temp), grid_lya, (xg, yg), method='cubic')
```

## Calculate actual emissivities at each gridpoint

### Emissivity from gas

Since we use 1 cm slabs, the emissivity is equivalent to the emergent intensity (intensity = 1cm * emissivity)

```python
lum = griddata((grid_dense, grid_temp), grid_lya, (np.log10(nH_sum), np.log10(temp_avg)))
lum = np.nan_to_num(lum, nan=-40)
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
plt.imshow(lum, cmap = 'jet', extent = [x.min(), x.max(), y.min(), y.max()], vmin=-39, vmax=-21, origin='lower')
plt.xlabel('Distance from Center [kpc]')
plt.ylabel('Distance from Center [kpc]')
plt.colorbar(label = r'log(intensity) [erg cm$^{{-2}}$ s$^{{-1}}$]')
plt.savefig('Simulated_Emission.png', dpi=500)
```

## Intensity from stars

### Convert into angular size

```python
arcsec_kpc = cosmo.arcsec_per_kpc_comoving(redshift)
```

```python
x_arcsec = x*u.kpc  * arcsec_kpc
y_arcsec = y*u.kpc * arcsec_kpc
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
plt.imshow(lum, cmap = 'jet', extent = [x_arcsec.min(), x_arcsec.max(), y_arcsec.min(), y_arcsec.max()], vmin=-32, vmax=-15.5, origin='lower')# -39 -21
plt.xlabel('Distance from Center [arcsec]')
plt.ylabel('Distance from Center [arcsec]')
plt.colorbar(label = r'log(intensity) [erg cm$^{{-2}}$ s$^{{-1}}$]')
plt.savefig('Simulated_Emission.png', dpi=500)
```

### Convert into surface brightness (intensity per steradian)

```python
arcsec_pp = (x_arcsec.max() - x_arcsec.min()) / n
```

```python
lum_abs = 10**lum
lum_abs = u.Quantity(lum_abs, u.erg / u.s / (u.cm**2))
```

```python
surface_brightness = (lum_abs / (arcsec_pp**2)).to(u.erg / u.s / (u.cm**2) / (u.arcsec**2))
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
plt.imshow(np.log10(surface_brightness.value), cmap = 'jet', vmin=-35, vmax=-16, extent = [x_arcsec.min(), x_arcsec.max(), y_arcsec.min(), y_arcsec.max()], origin='lower')
plt.xlabel('Distance from Center [arcsec]')
plt.ylabel('Distance from Center [arcsec]')
plt.colorbar(label = r'log(s. b.) [erg cm$^{{-2}}$ s$^{{-1}}$ arcsec$^{{-2}}$]')
plt.savefig('Simulated_Flux.png', dpi=500)
```

### Convolve with instrument FWHM

```python
fwhm = 6*u.arcsecond / arcsec_pp 
# number of pixels to blur by to achieve proper fwhm
```

```python
kernel = Gaussian2DKernel(x_stddev=fwhm)#, x_size=2*n, y_size=2*n)
```

```python
plt.imshow(kernel)
plt.colorbar()
```

```python
astropy_conv = convolve_fft(10**lum, kernel)
astropy_conv_plot = np.log10(astropy_conv)
```

```python
fig, ax = plt.subplots(figsize = (6, 4))
plt.imshow(astropy_conv_plot, cmap = 'jet', extent = [x_arcsec.min(), x_arcsec.max(), y_arcsec.min(), y_arcsec.max()], origin='lower')#, vmin=-40, vmax=-22)
plt.xlabel('Distance from Center (arcsec)')
plt.ylabel('Distance from Center (arcsec)')
plt.colorbar(label = r'log(flux) [erg cm$^{{-2}}$ s$^{{-1}}$ arcsec$^{{-2}}$]')
plt.savefig('Simulated_Observation.png', dpi=500)
```

```python
fb_fov = 20 * u.arcmin
fb_pix = 1000

fb_pix_scale = fb_fov / fb_pix

npix = ((x_arcsec.max() - x_arcsec.min()) / fb_pix_scale).si
```
