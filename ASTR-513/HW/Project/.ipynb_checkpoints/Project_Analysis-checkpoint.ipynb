```python
import gizmo_analysis as gizmo  # rename these packages for brevity
import utilities as ut  # rename these packages for brevity

import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.colors as cl
from scipy.stats import gaussian_kde
```

```python
gizmo.io
```

```python
simulation_directory = '/home/baparker/m12r_res7100/'
```

```python
part = gizmo.io.Read.read_snapshots('all', 'redshift', 0.702, simulation_directory)
```

```python
part['gas'].keys()
```

```python
gas_loc = part['gas']['position']
gas_temp = part['gas']['temperature']
gas_den = part['gas']['density']
```

```python
np.shape(gas_loc)
```

```python
scale = 100
x = gas_loc[::scale, 0]
y = gas_loc[::scale, 1]
z = gas_loc[::scale, 2]
temp = gas_temp[::scale]
dens = gas_den[::scale]
nH = part['gas'].prop('number.density')[::scale]
mass_frac = part['gas']['massfraction'][::scale, 0]
mass_frac_tot = np.sum(part['gas']['massfraction'][::scale, :], axis=1)
```

```python
ind = np.argsort(nH)

plt.scatter(y[ind], z[ind], c = nH[ind], s=0.1, norm=cl.LogNorm(), cmap = 'magma')
plt.xlabel('X (pc)')
plt.ylabel('Y (pc)')
plt.colorbar(label = 'Number Density')
# BP Plotting the number density of gas throughout the galaxy.
```

```python
ind = np.argsort(mass_frac)

plt.scatter(dens[ind], temp[ind], c = mass_frac[ind], s=0.1, norm=cl.LogNorm(), cmap='jet')
plt.xlabel('Density')
plt.ylabel('Temperature (K)')
plt.xscale('log')
plt.yscale('log')
plt.colorbar(label = 'Mass Fraction')
#plt.title('

# BP Plotting the phase diagram (density temperature) number density of all the gas particles.
```
